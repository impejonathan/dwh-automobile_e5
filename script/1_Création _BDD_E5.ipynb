Cellule 7 : üîç V√©rification compl√®te des tables cr√©√©es


```python
print("\nüîç V√âRIFICATION COMPL√àTE DU DATA WAREHOUSE")
print("=" * 70)

# ========================================
# V√©rification des tables
# ========================================
cursor.execute("""
SELECT 
    TABLE_NAME,
    CASE 
        WHEN TABLE_NAME LIKE 'DIM_%' THEN 'Dimension'
        WHEN TABLE_NAME LIKE 'FAIT_%' THEN 'Fait'
        ELSE 'Autre'
    END AS Type_Table
FROM INFORMATION_SCHEMA.TABLES 
WHERE TABLE_TYPE = 'BASE TABLE'
AND (TABLE_NAME LIKE 'DIM_%' OR TABLE_NAME LIKE 'FAIT_%')
ORDER BY Type_Table DESC, TABLE_NAME
""")

tables = cursor.fetchall()

print("\nüìã TABLES CR√â√âES DANS LE DATA WAREHOUSE :")
print("=" * 70)

dim_count = 0
fait_count = 0

for table in tables:
    if table[1] == 'Dimension':
        dim_count += 1
        print(f"  ‚úì {table[0]} (Dimension)")
    else:
        fait_count += 1
        print(f"  ‚úì {table[0]} (Fait)")

print("=" * 70)
print(f"Total : {len(tables)} tables ({dim_count} dimensions + {fait_count} faits)")

# ========================================
# V√©rification des contraintes de cl√©s √©trang√®res
# ========================================
print("\nüîó CONTRAINTES DE CL√âS √âTRANG√àRES :")
print("=" * 70)

cursor.execute("""
SELECT 
    tp.name AS 'Table',
    fk.name AS 'Contrainte_FK',
    cp.name AS 'Colonne',
    tr.name AS 'Table_Reference',
    cr.name AS 'Colonne_Reference'
FROM sys.foreign_keys AS fk
INNER JOIN sys.tables AS tp ON fk.parent_object_id = tp.object_id
INNER JOIN sys.tables AS tr ON fk.referenced_object_id = tr.object_id
INNER JOIN sys.foreign_key_columns AS fkc ON fk.object_id = fkc.constraint_object_id
INNER JOIN sys.columns AS cp ON fkc.parent_column_id = cp.column_id AND fkc.parent_object_id = cp.object_id
INNER JOIN sys.columns AS cr ON fkc.referenced_column_id = cr.column_id AND fkc.referenced_object_id = cr.object_id
WHERE tp.name LIKE 'FAIT_%'
ORDER BY tp.name, fk.name
""")

fks = cursor.fetchall()

for fk in fks:
    print(f"  ‚úì {fk[0]}.{fk[2]} ‚Üí {fk[3]}.{fk[4]} ({fk[1]})")

print("=" * 70)
print(f"Total : {len(fks)} contraintes FK")

# ========================================
# V√©rification des index cr√©√©s
# ========================================
print("\nüìä INDEX CR√â√âS POUR L'OPTIMISATION :")
print("=" * 70)

cursor.execute("""
SELECT 
    t.name AS 'Table',
    i.name AS 'Index',
    c.name AS 'Colonne',
    i.type_desc AS 'Type'
FROM sys.indexes i
INNER JOIN sys.tables t ON i.object_id = t.object_id
INNER JOIN sys.index_columns ic ON i.object_id = ic.object_id AND i.index_id = ic.index_id
INNER JOIN sys.columns c ON ic.object_id = c.object_id AND ic.column_id = c.column_id
WHERE (t.name LIKE 'DIM_%' OR t.name LIKE 'FAIT_%')
AND i.is_primary_key = 0
AND i.type_desc = 'NONCLUSTERED'
ORDER BY t.name, i.name
""")

indexes = cursor.fetchall()

# Compter les index uniques
unique_indexes = set()
for idx in indexes:
    unique_indexes.add(f"{idx[0]}.{idx[1]}")

print(f"Total : {len(unique_indexes)} index cr√©√©s")
print("\nD√©tail par table :")

current_table = ""
index_count_by_table = {}

for idx in indexes:
    if idx[0] != current_table:
        if current_table != "":
            print(f"  ‚Ä¢ {current_table} : {index_count_by_table[current_table]} index")
        current_table = idx[0]
        index_count_by_table[current_table] = 0
    
    if idx[1] not in [i[1] for i in indexes if i[0] == current_table][:index_count_by_table[current_table]]:
        index_count_by_table[current_table] += 1

# Afficher le dernier
if current_table:
    print(f"  ‚Ä¢ {current_table} : {index_count_by_table[current_table]} index")

print("=" * 70)

# ========================================
# Statistiques finales
# ========================================
print("\nüìä STATISTIQUES DU DATA WAREHOUSE :")
print("=" * 70)

cursor.execute("""
SELECT 
    t.name AS 'Table',
    p.rows AS 'Nombre_Lignes'
FROM sys.tables t
INNER JOIN sys.partitions p ON t.object_id = p.object_id
WHERE (t.name LIKE 'DIM_%' OR t.name LIKE 'FAIT_%')
AND p.index_id IN (0,1)
ORDER BY t.name
""")

stats = cursor.fetchall()

for stat in stats:
    print(f"  ‚Ä¢ {stat[0]} : {stat[1]} lignes")

print("=" * 70)
```
